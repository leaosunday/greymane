# 实现 Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1., 2., 3., 4., 5.])
y = np.array([1., 3., 2., 3., 5.])
```

```python
plt.scatter(x, y)
plt.axis([0, 6, 0, 6])
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
a = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
b = y_mean - a * x_mean
```

```python
a
```

```python
b
```

```python
y_hat = a * x + b
plt.scatter(x, y)
plt.plot(x, y_hat)
```

```python
x_predict = 6
y_predict = a * x_predict + b
y_predict
```
