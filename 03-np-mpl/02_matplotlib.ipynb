```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
siny = y.copy()
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red')
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')  # : -. -- -
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
plt.xlim(-2, 12)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
plt.ylim(0, 1.5)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
plt.axis([-1, 11, -2, 2])  # xlim, ylim
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
plt.xlabel('x axis')
plt.ylabel('y value')
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, color='red', linestyle='--', label='cos(x)')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()  # 
plt.title('Matplotlib Line Chart Example')
```

## Scatter Plot
